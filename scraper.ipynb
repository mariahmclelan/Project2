```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
```

```python
browser = Browser('chrome')
```

```python
url = "https://schools.texastribune.org/districts/"
browser.visit(url)
```

```python
html_contents = browser.html

districts_soup = soup(html_contents, 'html.parser')

districts_soup
```

```python
district_links = districts_soup.select('tbody>tr>td>a')
district_links
```

```python
district_urls = []
for a in district_links:
    district_urls.append(a['href'])
district_urls    
```

```python
#create an empty list to store information on districts
district_information = []
for links in district_urls:
    #click on link
    browser.links.find_by_href(links).click()
    #find all divs within textbox
    content = districts_soup.find_all('div', {'class':'intro-block'})
    #loop through each data metric div on page
    for metrics in content:
        data = metrics.select('div p')
        #create a list to hold page's metric data that clears on each page visit
        metric_data = []
        #loop through 
        for text in data:
            metric_data.append(text)
    #add all text data to list
    district_information.append(metric_data)
    #revert back to links page and reiterate on next district link
    browser.back()
    
```
