```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
pd.set_option('max_colwidth', 2000)
```

```python
browser = Browser('chrome')
```

```python
url = "https://schools.texastribune.org/districts/"
browser.visit(url)
```

```python
#view and store html contents of the initial site
html_contents = browser.html

districts_soup = soup(html_contents, 'html.parser')

districts_soup
```

```python
#target hyperlink hrefs in html
district_links = districts_soup.select('tbody>tr>td>a')
district_links
```

```python
#create a list of hyperlink hrefs
district_urls = []
for a in district_links:
    district_urls.append(a['href'])
district_urls    
```

```python
#create an empty list to store html frpm district pages
district_html = []
for links in district_urls:
    
    #click on link
    browser.links.find_by_href(links).click()
    
    page_html = browser.html

    district_soup = soup(page_html, 'html.parser')
    
    district_html.append(district_soup)
    
    #revert back to hyperlinks page and reiterate on next district hyperlink
    browser.back()
    
    
    
    
```

```python
#narrow html elemtents to just divs containing p tags for metrics
district_metrics = []
for pages in district_html:
    #tags = district_soup.select('main>div>section')
    tags = district_soup.find_all('section')
    p_tags = district_soup.select('div>p')
    district_metrics.append(tags)
    
```

```python
district_metrics[0]
```

```python
#reduce to just metrics and text from p tags
metric_data = []
for p in district_metrics:
    for element in p:
        if hasattr(element, 'text'):
            metric_data.append(element.text)
```

```python
metric_data
```

```python

#remove unwanted characters within list entries
for i in range(len(metric_data)):
    metric_data[i] = [re.sub(',', '',j) for j in metric_data[i]]
    metric_data[i] = [re.sub('\n', '',j) for j in metric_data[i]] 
metric_data
```

```python
#compress back into a clean and readable entry
clean_metrics = []
for entry in range(len(metric_data)):
    clean_entry = [str(item) for item in metric_data[entry]]
    clean_metrics.append("".join(clean_entry))
clean_metrics
```

```python
df_ready_metrics = []
for line in range(len(clean_metrics)):
    ready_entry = [str(line) for line in clean_metrics[line]]
    df_ready_metrics.append("".join(ready_entry))
df_ready_metrics    
```
