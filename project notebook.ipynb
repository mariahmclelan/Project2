```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
scores_finances  = pd.read_json('resources/scores_finances.json')
```

```python
schools  = pd.read_json('resources/texas-school.json')
```

```python
df = pd.DataFrame(scores_finances)

df2 = pd.DataFrame(schools)
```

```python
df = df.rename(columns=lambda x: x.replace(' ', '_'))
df
```

```python
df['Campus_Name'] = df['Campus_Name'].str.lower()
```

```python
df2.drop(labels=['CountyNumber', 'CountyName', 'RegionNumber', 'DistrictNumber',
       'DistrictType', 'DistrictStreetAddress', 'DistrictCity',
       'DistrictState', 'DistrictZip', 'DistrictSiteStreetAddress',
       'DistrictSiteCity', 'DistrictSiteState', 'DistrictSiteZip',
       'DistrictPhone', 'DistrictFax', 'DistrictEmailAddress',
       'DistrictWebPageAddress', 'DistrictSuperintendent',
       'DistrictEnrollmentasofOct2014', 'SchoolNumber',
       'InstructionType', 'CharterType', 'SchoolCity',
       'SchoolState', 'SchoolZip', 'SchoolSiteStreetAddress', 'SchoolSiteCity',
       'SchoolSiteState', 'SchoolSiteZip', 'SchoolPhone', 'SchoolFax',
       'SchoolEmailAddress', 'SchoolWebPageAddress', 'SchoolPrincipal',
       'GradeRange', 'SchoolEnrollmentasofOct2014', 'SchoolStatus',
       'SchoolStatusDate', 'UpdateDate', 'GeocodeQualityType','SchoolStreetAddress'], axis=1, inplace=True)
```

```python
df2.dropna()
```

```python
df2.rename(columns={'SchoolName': 'Campus_Name', 'DistrictName':'District_Name'}, inplace=True)
```

```python
df2['District_Name'] = df2['District_Name'].str.lower()
df2['Campus_Name'] = df2['Campus_Name'].str.lower()  
```

```python
df2
```

```python
school_info = pd.merge(df, df2, how = "left", on=['District_Name', 'Campus_Name'] )
```

```python
school_info = school_info.dropna()
```

```python
school_info
```

```python

# Define a function to determine marker size based on total operating revenue
def get_marker_size(revenue):
    if revenue < 5000000:
        return 5
    elif 5000000 <= revenue < 10000000:
        return 10
    elif 10000000 <= revenue < 50000000:
        return 15
    elif 50000000 <= revenue < 100000000:
        return 20
    elif 100000000 <= revenue < 500000000:
        return 25
    else:
        return 30

# Define a function to determine marker color based on student count based on UIL conferences 6A-1A
def get_marker_color(student_count):
    if student_count >= 2275:
        return 'red'
    elif 1315 <= student_count <= 2274:
        return 'purple'
    elif 545 <= student_count <= 1314:
        return 'blue'
    elif 254 <= student_count <= 544:
        return 'yellow'
    elif 105 <= student_count <= 253:
        return 'orange'
    else:
        return 'green'


map = folium.Map(location=[29.7604, -95.3698], zoom_start=6)
# Create a feature group for the markers
    
marker_group = folium.FeatureGroup(name='Markers')

# Iterate through the DataFrame to create markers with varying size and color
for idx, row in school_info.iterrows():
    size = get_marker_size(row['Total_Operating_Revenue'])
    color = get_marker_color(row['Student_Count'])
    popup_text = f"ISD Name: {row['District_Name']}<br>Revenue: ${row['Total_Operating_Revenue']:,}<br>Student Count: {row['Student_Count']:,}"
    marker = folium.CircleMarker(location=[row['Latitude'], row['Longitude']],
                                  radius=size,
                                  color=color,
                                  fill=True,
                                  fill_color=color,
                                  fill_opacity=0.5,
                                  popup=popup_text)
    marker.add_to(marker_group)

# Add the feature group to the map
marker_group.add_to(map)

# Add a layer control to the map
folium.LayerControl().add_to(map)

# Define legend HTML
legend_html = """
<script type="text/javascript">
    var legend = L.control({position: 'bottomleft'});

    legend.onAdd = function (map) {
        var div = L.DomUtil.create('div', 'info legend');
        div.innerHTML += '<b>Legend</b><br>';
        div.innerHTML += '<i style="background: red">&nbsp;&nbsp;</i> 6A: 2275 and above<br>';
        div.innerHTML += '<i style="background: purple">&nbsp;&nbsp;</i> 5A: 1315 - 2274<br>';
        div.innerHTML += '<i style="background: blue">&nbsp;&nbsp;</i> 4A: 545 - 1314<br>';
        div.innerHTML += '<i style="background: yellow">&nbsp;&nbsp;</i> 3A: 254 - 544<br>';
        div.innerHTML += '<i style="background: orange">&nbsp;&nbsp;</i> 2A: 105 - 253<br>';
        div.innerHTML += '<i style="background: green">&nbsp;&nbsp;</i> 1A: Below 105<br>';
        return div;
    };

    legend.addTo(map);
</script>
"""

# Add the legend HTML to the map's HTML
map.get_root().html.add_child(folium.Element(legend_html))
# Save the map
map.save('map_with_markers_revenue_size_and_student_count_color_updated.html')
```
